```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import csv
%matplotlib inline
```

```python
files = ['driving_log-3lapclockwise.csv',
         'driving_log-3lapcounterclockwise.csv',
         'driving_log-recoveryclockwise.csv',
         'driving_log-recovery-counter-clockwise.csv']
```

```python
from random import random
data = []
for f in files:
    with open(f) as csvfile:
        reader = csv.reader(csvfile)
        for line in reader:
            steering = float(line[3])
            if steering == 0.0 and random() <= 0.5:
                continue
            data.append(float(line[3]))
```

```python
df = pd.Series(data)
```
